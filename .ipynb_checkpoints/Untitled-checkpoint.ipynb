```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data7 = pd.read_csv("file.csv")
```

```python
data7 = data7.dropna()
```

```python
data7.info()
```

```python
data7.duplicated().value_counts()
```

```python
data8 = data7[:]
```

```python
data7.CustomerID.nunique()
```

```python
data7.drop(['Transaction_ID','Product_SKU','Transaction_Date','Date'],axis=1,inplace=True)
```

```python
data7=data7.drop_duplicates()
```

```python
data7['Total Prices']=data7.Avg_Price+data7.Delivery_Charges
```

```python
val1=data7['Total Prices']
```

```python
val1=data7[['CustomerID','Total Prices']]
```

```python
x=val1.CustomerID.unique()[:40]
y=val1['Total Prices'].unique()[:40]
```

```python
fig,axis=plt.subplots(figsize=(12,12))

val1=val1.sort_values(by='Total Prices',ascending=False)
sb.barplot(x=x,y=y,palette='hot',ax=axis)
plt.xticks(rotation=90)
plt.title("Maximum Price Customer Id")
```

```python
val2=data7.CustomerID.value_counts().sort_values(ascending=False).head(30)
sb.barplot(x=val2.index,y=val2,palette='rainbow')

plt.xticks(rotation=90)
plt.ylabel('No. of Purchases')
plt.title('Top 30 Popular Customer ID with Purchase Count')
```

```python
val3=data7.Gender.value_counts()
plt.pie(val3,labels=['Female','Male'],autopct="%1.1f%%")
plt.axis('equal')
plt.title('Male - Female Purchase Comparision')
sb.set(style='white')
```

```python
val5=data7.Avg_Price
sb.boxplot(val5,palette='hls')
plt.title('Average Price Anaysis')
plt.xlabel('Average Price')
plt.xticks([0],['Central Tendency'])
plt.ylabel('Prices')
```

```python
data7['Total_Spend']=data7['Offline_Spend']+data7['Online_Spend']
new=data7.groupby('Month').sum()
```

```python
import calendar
mon=list(calendar.month_name)[1:]
```

```python
sb.lineplot(y=new.Total_Spend,x=new.index,color='g')
plt.xticks(new.index,mon,rotation=60)
plt.title("Total Spend With Month")
```

```python
sb.lineplot(y=new.Online_Spend,x=new.index,color='pink')
plt.xticks(new.index,mon,rotation=60)
plt.title("Total Online Spend With Month")
```

```python
sb.lineplot(y=new.Offline_Spend,x=new.index,color='orange')
plt.xticks(new.index,mon,rotation=60)
plt.title("Total Offline Spend With Month")
```

```python
delivery=data7[['Delivery_Charges','Month']].groupby('Month').sum()
sb.barplot(data=delivery,x=delivery.index,y='Delivery_Charges',palette='coolwarm')
plt.xticks(delivery.index,mon,rotation=60)
plt.title("Total Delivery Charges per Month")
```

```python
sb.scatterplot(data=data7,x='Avg_Price',y='Delivery_Charges',hue="Month",palette='cubehelix')
plt.title("Delivery Charges Vs Average Price By Two Month Separation")
```

```python
new1=data7.groupby('Product_Category').size()
plt.pie(new1,labels=new1.index,autopct="%1.2f%%",startangle=90,explode=[0.8 if i%2==1 else 0 for i in range(len(new1))])
plt.title("Product Category Frequency Analysis")
```

```python
import matplotlib.pyplot as plt


new1 = data7.groupby('Product_Category').size()


fig, ax = plt.subplots()


new1.sort_values().plot(kind='barh', color='skyblue', edgecolor='black', ax=ax)


for index, value in enumerate(new1.sort_values()):
    ax.text(value, index, f'{value/sum(new1)*100:.2f}%', va='center')

# Set labels and title
ax.set_xlabel('Frequency')
ax.set_ylabel('Product Category')
plt.title("Product Category Frequency Analysis")

# Show the plot
plt.show()
```

```python
sb.violinplot(data=new,x='Quantity')
plt.title("Total Quantity Per Month Analysis")
```
